```python
import matplotlib.pyplot as plt
import numpy as np
from wordcloud import WordCloud
```

```python
with open("/content/text.txt", "r", encoding="UTF-8") as file:
    contents = file.read()
```

```python
text=" ".join(contents.split())
print(text)
```

```python
word_cloud=WordCloud(
    background_color="black",
    colormap="viridis",
    contour_color="steelblue",
    contour_width=2,
    random_state=42
)

word_cloud.generate(text)
plt.figure(figsize=(10,10))
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

**By using a Mask**

```python
from PIL import Image
Mask=np.array(Image.open("/content/pngwing.com.png"))
```

```python
word_cloud=WordCloud(
    mask=Mask,
    background_color="black",
    colormap="viridis",
    contour_color="steelblue",
    contour_width=2
)

word_cloud.generate(text)
plt.figure(figsize=(10,10))
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
